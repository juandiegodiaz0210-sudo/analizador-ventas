# Analizador de Ventas para Pequeños Negocios 
##### Herramienta en Python que analiza datos de ventas y genera reportes con insights clave para la toma de decisiones.

## Introducción
##### Este proyecto consiste en el desarrollo de un analizador de ventas, haciendo uso de Phyton. Tiene como objetivo procesar datos, limpiarlos y facilitar información útil, como los productos más vendidos en la semana, mes o años o los que tienen un periodo de mayor rendimiento. 

##### Está pensado para apoyar la toma de decisiones en pequeños negocios mediante el análisis de datos.

## Carga de Datos
##### En esta sección se realiza la carga de los datos de ventas desde un archivo CSV. Estos datos serán utilizados posteriormente para su limpieza, transformación y análisis.

```python
import pandas as pd

# Solicitar ruta del archivo al usuario
archivo1 = input('Ingresa la ruta del primer archivo CSV: ')
archivo2 = input('Ingresa la ruta del segundo archivo CSV: ')


try:
    # Cargar Datos
    df1 = pd.read_csv(archivo1)
    df2 = pd.read_csv(archivo2)

    # Unir los datos
    df = pd.concat([df1, df2])
    
    print('Datos cargados y concatenados correctamente')
    print(df.head())

except FileNotFoundError:
    print('Error: uno de los archivos no se encontró')

except Exception as e:
    print('Ocurrió un error:', e)
```

## Limpieza de Datos
##### En esta sección se realiza la limpieza de los datos con el objetivo de eliminar valores nulos, corregir formatos y asegurar la calidad de la información antes del análisis.

```python
# Ver valores nulos
print(df.isnull().sum())

# Eliminar datos nulos
df = df.dropna()

# Eliminar duplicados
df = df.drop_duplicates()

# Convertir columna de fecha 
df['Fecha'] = df['Fecha'] = pd.to_datetime(df['Fecha'])

# Crear columna de mes
df['Mes'] = df['Fecha'].dt.month_name

# Verificar estructura 
print(df.info())
```

## Análisis de Datos

##### En esta sección se realizará el análisis de los datos con el objetivo de identificar patrones, tendencias y puntos clave como los productos más vendidos y los periodos con mayores ventas.

## Ventas por producto

```python

ventas_por_producto = df.groupby('Producto')['Total Venta'].sum()
print(ventas_por_producto)
```

## Producto más vendido

```python
print('Producto más vendido:', ventas_por_producto.idxmax())
```

## Ventas por mes

```python
ventas_por_mes = df.groupby('Mes')['Total Venta'].sum()
print(ventas_por_mes)
```

## Mejor mes

```python
print('Mes con más ventas:', ventas_por_mes.idxmax())
```

# Gráficos
### Ventas por producto

```python
import matplotlib.pyplot as plt

ventas_por_producto.plot(kind='bar')
plt.title("Ventas por Producto")
plt.xlabel("Producto")
plt.ylabel("Ventas")
plt.show()
```

## Ventas por mes 

```python
ventas_por_mes.plot(kind='line')
plt.title("Ventas por Mes")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.show()
```

# Resultados 
#### A partir de este análisis realizado, se identificó que el producto más vendido es x, mientras que el mes con mayores ventas fue Y.
#### Estos resultados nos permiten entender mejor el comportamiento de las ventas y pueden ser utilizados para mejorar la toma de decisiones en un negocio.
